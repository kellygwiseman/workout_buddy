```python
from scipy import signal

# Conditions
cond1 = np.abs(correls.ix[:, 'gyroRotationX', 'gyroRotationZ']) > 0.6
cond2 = np.abs(correls.ix[:, 'accelerometerAccelerationX', 'accelerometerAccelerationY']) > 0.5
cond3 = np.abs(correls.ix[:, 'accelerometerAccelerationX', 'accelerometerAccelerationZ']) > 0.5
cond4 = np.abs(correls.ix[:, 'motionQuaternionY', 'motionQuaternionZ']) > 0.7
cond5 = np.abs(correls.ix[:, 'motionQuaternionX', 'motionQuaternionY']) > 0.7
cond6 = np.abs(correls.ix[:, 'motionQuaternionX', 'motionQuaternionW']) > 0.7
# don't use the first 3 and last 2 seconds
cond7 = df_filt.index > 60 
cond8 = df_filt.index < (df_filt.shape[0] - 40) 
pushup_window = df_filt[cond1 & cond2 & cond4 & cond5 & cond7 & cond8]

widths = np.arange(6,14)
gap = 1.0
max_dist = np.linspace(4.0,4.0,df.shape[0])
length = 4
min_snr = 1.0
#peakind = gr.plot_pushups(df_filt, pushup_window, widths, max_dist, gap, freq, sample)
#peakind = plot_pushups1(df_num, pushup_window, widths, length, min_snr, freq, sample)
peakind = plot_pushups2(df_filt, pushup_window, widths, max_dist, gap, freq, sample)
print peakind
print 'number of pushups = ', len(peakind)
```

```python
def plot_pushups1(data, pushup_window, widths, length, min_snr, freq, sample):
    time = data.index.values / freq
    pushup_start = pushup_window.index[0] / freq
    pushup_time = pushup_window.index.values / freq
    
    # calculate push-up reps
    peakind = signal.find_peaks_cwt(pushup_window['motionPitch'], widths = widths,
                                    min_length = length, min_snr = min_snr)
    peakind = [x / freq for x in peakind]
    
    # Plot complete time series and push-up duration overlay
    fig1 = plt.figure()
    plt.plot(time, data.motionPitch, label='Raw data')
    plt.plot(pushup_time, pushup_window['motionPitch'], label='Push-up duration')
    
    # Plot push-ups
    plt.scatter(pushup_start + peakind,np.linspace(0.0,0.0,num=len(peakind)), color='r',marker='x', lw=2, label='Push-ups')
    plt.title('Push-up Repetitions')
    plt.ylabel('Pitch (Radians)')
    plt.xlabel('Time (Seconds)')
    plt.xlim(0,time[-1])
    ymin, ymax = plt.ylim()
    #plt.ylim((ymin-0.2), ymax)
    plt.ylim(-0.75, 1.75)
    plt.legend(loc='lower center', ncol=3, bbox_to_anchor=(0.5, 0.0), frameon=False, columnspacing=1, borderpad=0.1)
    plt.savefig('../figures/pu_reps-'+sample+'.png')
    return peakind;
```

```python
def plot_pushups2(data, pushup_window, widths, max_dist, gap_thresh, freq, sample):
    time = data.index.values / freq
    pushup_start = pushup_window.index[0] / freq
    pushup_time = pushup_window.index.values / freq
    
    # calculate push-up reps
    peakind = signal.find_peaks_cwt(pushup_window['motionPitch'], widths = widths,
                                    max_distances = max_dist, gap_thresh = gap_thresh)
    peakind = [x / freq for x in peakind]
    
    # Plot complete time series and push-up duration overlay
    fig1 = plt.figure()
    plt.plot(time, data.motionPitch, label='Raw data')
    plt.plot(pushup_time, pushup_window['motionPitch'], label='Push-up duration')
    
    # Plot push-ups
    plt.scatter(pushup_start + peakind,np.linspace(0.0,0.0,num=len(peakind)), color='r',marker='x', lw=2, label='Push-ups')
    plt.title('Push-up Repetitions')
    plt.ylabel('Pitch (Radians)')
    plt.xlabel('Time (Seconds)')
    plt.xlim(0,time[-1])
    ymin, ymax = plt.ylim()
    #plt.ylim((ymin-0.2), ymax)
    plt.ylim(-1.0, 1.0)
    plt.legend(loc='lower center', ncol=3, bbox_to_anchor=(0.5, 0.0), frameon=False, columnspacing=1, borderpad=0.1)
    plt.savefig('../figures/pu_reps-'+sample+'.png')
    return peakind;
```
