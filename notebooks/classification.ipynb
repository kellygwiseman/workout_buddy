```python
import sys
sys.path.append(r'../code')
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split, StratifiedShuffleSplit
from sklearn.preprocessing import scale
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../processed/processed_all_pushup.csv')
df.drop(['Unnamed: 0','user_id','hertz','count','Pcount'], axis=1, inplace=True)
df.head(10)
```

```python
label = df.pop('form')
label = label.map({'excellent':1, 'good':1,'ok':0})
df['exercise'] = df['exercise'].map({'pushup':1, 'Kpushup':0})
df['stance'] = df['stance'].map({'normal':0,'narrow':1,'inward':2})
df['height (inches)'] = scale(df['height (inches)'])
df['avg_amp'] = scale(df['avg_amp'])
df['avg_dur'] = scale(df['avg_dur'])
df.head(5)
X = df.values
```

```python
feature_names = df.columns.values
```

```python
sss = StratifiedShuffleSplit(label, n_iter = 10, test_size = 0.3, random_state=100)
```

```python
acc_list = []
recall_list = []
prec_list = []
for train_index, test_index in sss:
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = label[train_index], label[test_index]
    rf = RandomForestClassifier(n_estimators=100, max_features=3, max_depth=2)
    rf.fit(X_train[:,4:], y_train)
    y_pred = rf.predict(X_test[:,4:])
    y_prob = rf.predict_proba(X_test[:,4:])
    #print rf.classes_
    #print y_prob, X_test[:,:2]
    acc = metrics.accuracy_score(y_test, y_pred)
    acc_list.append(acc)
    #print 'accuracy score for default parameters', acc
    recall = metrics.recall_score(y_test, y_pred)
    recall_list.append(recall)
    #print 'recall score for default parameters', recall
    precision = metrics.precision_score(y_test, y_pred)
    prec_list.append(precision)
    #print 'precision score for default parameters', precision
    cm = metrics.confusion_matrix(y_test, y_pred)
    print(cm)

    # Show confusion matrix in a separate window
    #plt.matshow(cm)
    #plt.title('Confusion matrix')
    #plt.colorbar()
    #plt.ylabel('True label')
    #plt.xlabel('Predicted label')
    #plt.show()
    
    #fi = rf.feature_importances_
    # sorted in ascending order
    #fi_index = np.argsort(fi)
    #fi_index += 4
    #print feature_names[fi_index]
print 'average accuracy and std:', np.mean(acc_list), np.std(acc_list)
print 'average precison and std:', np.mean(prec_list), np.std(prec_list)
print 'average recall and std:', np.mean(recall_list), np.std(recall_list)
```
