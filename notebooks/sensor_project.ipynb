# Pushup identification and counting algorithm

```python
# look at correlation between features
features = ['accelerometerAccelerationX','accelerometerAccelerationY','accelerometerAccelerationZ',
                       'gyroRotationX','gyroRotationY','gyroRotationZ','motionYaw','motionRoll','motionPitch']
ip6_gpu_num = ip6_gpu[features]
correls = pd.rolling_corr(ip6_gpu_num, 25)
correls.ix[:, 'gyroRotationX', 'gyroRotationZ'].plot()
correls.ix[:, 'motionYaw', 'motionPitch'].plot()
correls.ix[:, 'accelerometerAccelerationX', 'accelerometerAccelerationY'].plot()
```

```python
cond1 = correls.ix[:, 'gyroRotationX', 'gyroRotationZ']>0.8
cond2 = correls.ix[:, 'motionYaw', 'motionPitch'] > 0.8
push_window = ip6_gpu_num[cond1 & cond2]
push_start = push_window.index[0]
ip6_gpu_num['motionPitch'].plot()
push_window['motionPitch'].plot()
#push_window['motionYaw'].plot()
widths = np.arange(4,8)
peakind = signal.find_peaks_cwt(push_window['motionPitch'], widths)
plt.scatter(push_start + peakind,np.linspace(0.0,0.0,num=len(peakind)), color='r')
print peakind
print 'number of pushups = ', len(peakind)
```

```python
# look at correlation between features
features = ['accelerometerAccelerationX','accelerometerAccelerationY','accelerometerAccelerationZ',
                       'gyroRotationX','gyroRotationY','gyroRotationZ','motionYaw','motionRoll','motionPitch']
ip6_bpu_num = ip6_bpu[features]
correls = pd.rolling_corr(ip6_bpu_num, 25)
correls.ix[:, 'gyroRotationX', 'gyroRotationZ'].plot()
correls.ix[:, 'motionYaw', 'motionPitch'].plot()
correls.ix[:, 'accelerometerAccelerationX', 'accelerometerAccelerationY'].plot()

figure = plt.figure()
cond1 = correls.ix[:, 'gyroRotationX', 'gyroRotationZ']>0.8
cond2 = correls.ix[:, 'motionYaw', 'motionPitch'] > 0.8
cond3 = correls.ix[:, 'accelerometerAccelerationX', 'accelerometerAccelerationY'] > 0.8
push_window = ip6_bpu_num[cond1 & cond2]
push_start = push_window.index[0]
ip6_bpu_num['motionPitch'].plot()
push_window['motionPitch'].plot()
#push_window['motionYaw'].plot()
widths = np.arange(4,8)
peakind = signal.find_peaks_cwt(push_window['motionPitch'], widths)
plt.scatter(push_start + peakind,np.linspace(0.0,0.0,num=len(peakind)), color='r')
print peakind
print 'number of pushups = ', len(peakind)
```
