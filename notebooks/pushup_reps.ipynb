```python
import sys
sys.path.append(r'../code')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import graphs as gr
from sklearn.cross_validation import StratifiedShuffleSplit
from sklearn import metrics
from matplotlib import rcParams
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split, StratifiedShuffleSplit
from sklearn.preprocessing import scale
from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV
from sklearn.preprocessing import scale
from dtw import *
%matplotlib inline
```

## Load the data

```python
data_arr = np.load('../processed/pushup_pitch_rep_metrics.npy')
label = data_arr[:,4]
X = data_arr[:,2:4].astype(float)
label[label =='excellent'] = 1
label[label == 'good'] = 1
label[label == 'ok'] = 0
label = label.astype(int)
```

## Random Forest

```python
acc_list = []
recall_list = []
prec_list = []
sss = StratifiedShuffleSplit(label, n_iter = 10, test_size = 0.3, random_state=0)
for train_index, test_index in sss:
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = label[train_index], label[test_index]
    rf = RandomForestClassifier(n_estimators=50, max_features=2, max_depth=2)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    y_prob = rf.predict_proba(X_test)
    #print rf.classes_
    #print y_prob, X_test[:,:2]
    acc = metrics.accuracy_score(y_test, y_pred)
    acc_list.append(acc)
    #print 'accuracy score for default parameters', acc
    recall = metrics.recall_score(y_test, y_pred)
    recall_list.append(recall)
    #print 'recall score for default parameters', recall
    precision = metrics.precision_score(y_test, y_pred)
    prec_list.append(precision)
    #print 'precision score for default parameters', precision
    cm = metrics.confusion_matrix(y_test, y_pred)
    print(cm)

    # Show confusion matrix in a separate window
    #plt.matshow(cm)
    #plt.title('Confusion matrix')
    #plt.colorbar()
    #plt.ylabel('True label')
    #plt.xlabel('Predicted label')
    #plt.show()
    
    #fi = rf.feature_importances_
    #sorted in ascending order
    #fi_index = np.argsort(fi)
    #fi_index += 4
    #print feature_names[fi_index]
print 'average accuracy and std:', np.mean(acc_list), np.std(acc_list)
print 'average precison and std:', np.mean(prec_list), np.std(prec_list)
print 'average recall and std:', np.mean(recall_list), np.std(recall_list)
```

## Clustering

```python
# clustering
from sklearn.cluster import KMeans
label = data_arr[:,4]
X = data_arr[:,:4].astype(float)
X[:,1] = scale(X[:,1])
X[:,2] = scale(X[:,2])
X[:,3] = scale(X[:,3])
label[label =='excellent'] = 2
label[label == 'good'] = 1
label[label == 'ok'] = 0
label = label.astype(int)
features = ['Female','Height','Amplitude','Duration']
k = 3
km = KMeans(n_clusters=k)
km.fit_transform(X)
centroids = km.cluster_centers_
c_labels = km.labels_
```

```python
def plot_clusters(feature_names, X, clusters):
    '''Inputs:  
            feature_names: list of features
            X: (s x f) data array, s is number of samples and f is number of features
            clusters: (s,) array, s is number of samples
                
        Outputs: 
            3D plot of data, colored by cluster assignment'''
    
    # calculate number of clusters
    k = len(np.unique(clusters))
    
    fig = plt.figure(figsize=(14,6))
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax.scatter(X[:, 2], X[:, 3], X[:, 0], c=clusters)
    ax.set_title(str(k)+' Clusters')
    ax.set_xlabel(feature_names[2])
    #ax.w_xaxis.set_ticklabels([])
    ax.set_ylabel(feature_names[3])
    #ax.w_yaxis.set_ticklabels([])
    ax.set_zlabel(feature_names[0])
    #ax.w_zaxis.set_ticklabels([])

    ax = fig.add_subplot(1, 2, 2, projection='3d')
    ax.scatter(X[:, 2], X[:, 3], X[:, 1], c=clusters)
    ax.set_title(str(k)+' Clusters')
    ax.set_xlabel(feature_names[2])
    #ax.w_xaxis.set_ticklabels([])
    ax.set_ylabel(feature_names[3])
    #ax.w_yaxis.set_ticklabels([])
    ax.set_zlabel(feature_names[1])
    #ax.w_zaxis.set_ticklabels([])
```
